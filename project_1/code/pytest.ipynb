```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

def franke(x:np.ndarray, y:np.ndarray) -> np.ndarray:
  term1 = 0.75 * np.exp(-0.25*(9 * x - 2)**2 - 0.25*(9 * y - 2)**2)
  term2 = 0.75 * np.exp(-(9 * x + 1)**2 / 49. - 0.1*(9 * y + 1)**2)
  term3 = 0.5 * np.exp(-0.25*(9 * x - 7)**2 - 0.25*(9 * y - 3)**2)
  term4 = -0.2 * np.exp(-(9 * x - 4)**2 - (9 * y - 7)**2)
  return term1 + term2 + term3 + term4

def franke_training_data(
  n: int, 
  noise_amplitude: float=0.1, 
  random_inputs: bool=False
) -> tuple[np.ndarray, np.ndarray]:
  if random_inputs:
    x_1 = np.random.rand(n)
    x_2 = np.random.rand(n)
  else:
    x_1 = np.linspace(0, 1, n)
    x_2 = np.linspace(0, 1, n)
  X_1, X_2 = np.meshgrid(x_1, x_2)
  Y = franke(x_1, x_2)

  x_1, x_2 = X_1.ravel(), X_2.ravel()
  y = Y.ravel()
  print(y.shape)

  X = np.vstack((x_1, x_2)).T
  return X, y

samples = 10
max_degree = 10
trials = 100

testerror = np.zeros(max_degree)
trainingerror = np.zeros(max_degree)

x, y = franke_training_data(samples, noise_amplitude=0.1, random_inputs=True)
x_scaled = StandardScaler().fit_transform(x)

for degree in range(1, max_degree):

  x_poly = PolynomialFeatures(degree=degree).fit_transform(x_scaled)

  for trail in range(0, trials):
    x_train, x_test, y_train, y_test = train_test_split(x_poly, y, train_size=0.8)

    X_train = PolynomialFeatures(degree=5).fit_transform(x_train)
    X_test = PolynomialFeatures(degree=5).fit_transform(x_test)

    model = LinearRegression(fit_intercept=False).fit(x_train, y_train)
  y_pred_train = model.predict(x_train)
  y_pred_test = model.predict(x_test)

degrees = np.arange(0, max_degree)
plt.plot(degrees, np.log10(trainingerror), label='Training Error')
plt.plot(degrees, np.log10(testerror), label='Test Error')
plt.xlabel('Polynomial degree')
plt.ylabel('log10[MSE]')
plt.legend()
plt.show()
```
